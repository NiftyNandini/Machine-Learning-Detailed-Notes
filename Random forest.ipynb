## Random Forest Algorithm

- **Definition**: Random Forest is a popular machine learning algorithm under the **supervised learning technique**.
  - It is suitable for both **Classification** and **Regression** problems.
  - It relies on **ensemble learning**, combining multiple classifiers to enhance the model's performance.


- **Key Concept**:
  - As the name suggests, **"Random Forest is a classifier that contains a number of decision trees on various subsets of the given dataset and takes the average to improve the predictive accuracy of that dataset."**
  - Predictions from each tree are aggregated, and the final output is determined by:
    - **Majority voting** (for classification)
    - **Averaging** (for regression)
  - This ensemble approach boosts the predictive accuracy and robustness of the model.

**The greater number of trees in the forest leads to higher accuracy and prevents the problem of overfitting.**

## Assumptions for Random Forest

- Since Random Forest uses multiple decision trees, there are key assumptions to ensure accuracy:
  1. There should be **actual values** in the feature variables of the dataset for accurate prediction instead of guesses.
  2. Predictions from each tree should have **low correlations** to ensure diverse and reliable output.

## Why Use Random Forest?

- **Efficiency**: Takes less training time compared to other algorithms.
- **High Accuracy**: Delivers accurate results even with large datasets.
- **Handles Missing Data**: Maintains accuracy even if a significant portion of data is missing.

## How Does the Random Forest Algorithm Work?

- Random Forest works in two phases:
  1. **Creating the Random Forest**: Combines multiple decision trees.
  2. **Making Predictions**: Uses the combined trees for prediction.

### Step-by-Step Process:

1. **Step 1**: Select random **K data points** from the training set.
2. **Step 2**: Build **decision trees** using the selected subsets.
3. **Step 3**: Choose the number **N** for the total decision trees to build.
4. **Step 4**: Repeat **Steps 1 & 2** to create N decision trees.
5. **Step 5**: For new data points, find predictions from each decision tree and assign the category based on **majority voting**.


## Example: Random Forest Algorithm

### Scenario:
- Imagine a dataset containing images of different **fruits**.
- The goal is to classify these images using the **Random Forest classifier**.


#### Step-by-Step Process:
- A dataset with fruit images is split into **subsets**, each assigned to a decision tree.
- Each decision tree is trained to classify fruits using its data subset.
- For a **new fruit image**, each tree makes a prediction.
- The final classification is based on the **majority vote** from all trees.

#### Key Takeaway:
- Multiple decision trees combine to make a more accurate prediction, minimizing errors.


### Applications of Random Forest:

There are mainly four sectors where Random forest mostly used:

- **Banking:** Banking sector mostly uses this algorithm for the identification of loan risk.
- **Medicine:** With the help of this algorithm, disease trends and risks of the disease can be identified.
- **Land Use:** We can identify the areas of similar land use by this algorithm.
- **Marketing:** Marketing trends can be identified using this algorithm.

### Advantages of Random Forest

- Random Forest is capable of performing both Classification and Regression tasks.
- It is capable of handling large datasets with high dimensionality.
- It enhances the accuracy of the model and prevents the overfitting issue.

### Disadvantages of Random Forest

Although random forest can be used for both classification and regression tasks, it is not more suitable for Regression tasks.
