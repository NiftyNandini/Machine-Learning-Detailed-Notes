# STATISTICS :

Statistics is a branch of mathematics that is responsible for collecting, analyzing, interpreting, and presenting numerical data. It encompasses a wide array of methods and techniques used to summarize and make sense of complex datasets.

## 1.MEAN:

Definition:The average of a dataset, calculated by adding all values and dividing by the count.

Formula: 



$$\mu = \frac{\text{Sum of values}}{\text{Number of values}}$$
 

## 2.MEDIAN 

The middle value in a sorted dataset. 
For an even count of values, it’s the average of the two central values.

For odd number of data points:

**Median**: 
$$ \text{Median} = \frac{(n + 1)}{2} $$


for even:

Median =Average of (n/2) value and its next value

## 3.MODE

The value that appears most frequently in a dataset.

Use: Best for categorical data and helps identify the most common value.


## 4. Variance

Variance is a statistical measurement of how far each number in a data set is from the mean, and from every other number in the set. 

Formula :

**Variance**:
$$ \sigma^2 = \frac{\Sigma (x - \mu)^2}{N-1} $$

 
Use: Larger variance means data points are spread out; lower variance means they are closer to the mean.


## 5. Standard Deviation (SD)

The square root of variance, giving the average distance of data points from the mean.

Formula:
    
SD : $ σ = √ σ^2 $
    
Use: Useful for understanding data dispersion in the same units as the original data.

## 6. Covariance

Covariance measures the degree to which two variables change together.

Formula: 

**Covariance**:
$$ \text{Cov}(X, Y) = \frac{\Sigma (X_i - \bar{X})(Y_i - \bar{Y})}{n-1} $$

Use: A positive value indicates variables move together, and a negative value indicates they move inversely.

## 7. Correlation

Correlation measures the strength and direction of the linear relationship between two variables.

Formula:
    
**Pearson Correlation Coefficient**:
$$ \rho(X, Y) = \frac{\text{Cov}(X, Y)}{\sigma_X \sigma_Y} $$

Range: -1 to +1, where +1 is a perfect positive correlation it indicates a direct relationship, -1 is a perfect negative correlation it implies an inverse relationship, and 0 means no linear relationship.

## 8. Random Variable

A variable whose values depend on the outcomes of a random phenomenon.


Types: Discrete (finite values, e.g., dice rolls) and Continuous (infinite values within a range, e.g., height).

## 9. Percentile

A measure that indicates the value below which a given percentage of data points fall.

Use: Common in test scores (e.g., 90th percentile means 90% of values are below this score).


## 10. Skewness

Definition: A measure of asymmetry in a distribution.

Normal Distribution: Symmetrical, mean = median = mode.

Right-Skew (Positive Skew): Tail is longer on the right; mean > median.

Left-Skew (Negative Skew): Tail is longer on the left; mean < median.

<ul>
  <li>
    <strong>Right Skew:</strong>
    <ul>
      <li>Also known as positive skewness.</li>
      <li>Characteristics:</li>
      <ul>
        <li>Longer or fatter tail on the right-hand side (upper tail).</li>
        <li>More extreme values on the right side.</li>
        <li>Mean > Median.</li>
        <li>Indicates a distribution that is skewed towards the left.</li>
      </ul>
    </ul>
  </li>

  <li>
    <strong>Left Skew:</strong>
    <ul>
      <li>Also known as negative skewness.</li>
      <li>Characteristics:</li>
      <ul>
        <li>Longer or fatter tail on the left-hand side (lower tail).</li>
        <li>More extreme values on the left side.</li>
        <li>Mean < Median.</li>
        <li>Indicates a distribution that is skewed towards the right.</li>
      </ul>
    </ul>
  </li>

  <li>
    <strong>Zero Skew:</strong>
    <ul>
      <li>Also known as symmetrical distribution.</li>
      <li>Characteristics:</li>
      <ul>
        <li>Symmetric distribution.</li>
        <li>Left and right sides are mirror images of each other.</li>
        <li>Mean = Median.</li>
        <li>Indicates a distribution with no skewness.</li>
      </ul>
    </ul>
  </li>
</ul>


# Interquartile Range Formula

The difference between the upper and lower quartile is known as the interquartile range. The formula for the interquartile range is given below

Interquartile range = Upper Quartile – Lower Quartile = Q­3 – Q­1

  <li>Given dataset: 77, 85, 92, 64, 78, 95, 82</li>

  <li>Step 2: Count the values:
    <ul>
      <li>There are 7 values, which is an odd count.</li>
      <li>The median (middle value) is 82.</li>
    </ul>
  </li>
  <li>Step 3: Divide the data into two halves:
    <ul>
      <li>Lower half: 64, 77, 78</li>
      <li>Upper half: 85, 92, 95</li>
    </ul>
  </li>
  <li>Step 4: Identify Q1 and Q3:
    <ul>
      <li>Q1 (median of lower half) = 77</li>
      <li>Q3 (median of upper half) = 92</li>
    </ul>
  </li>
  <li>Step 5: Calculate the Interquartile Range (IQR):
    <ul>
      <li>IQR = Q3 - Q1 = 92 - 77 = 15</li>
    </ul>
  </li>
</ul>


## Outliers: 

In the interquartile range (IQR) method, an outlier is a data point that is more than 1.5 times the IQR above the third quartile or below the first quartile:

low_lim = Q1 - 1.5 * IQR

up_lim = Q3 + 1.5 * IQR

```python
import numpy as np

# Sample dataset
data = [77, 85, 92, 64, 78, 95, 82]

# Step 1: Arrange data in ascending order
data_sorted = sorted(data)

# Step 2: Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = np.percentile(data_sorted, 25)
Q3 = np.percentile(data_sorted, 75)

# Step 3: Calculate IQR
IQR = Q3 - Q1

# Step 4: Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Step 5: Identify outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]

# Display results
print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Outliers:", outliers)
```

```python
import numpy as np

# Sample dataset with potential outliers
data = [10, 12, 14, 15, 18, 19, 21, 22, 100]  
# Sort the data
data_sorted = sorted(data)

# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = np.percentile(data_sorted, 25)
Q3 = np.percentile(data_sorted, 75)

# Calculate IQR
IQR = Q3 - Q1

# Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]

# Display results
print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Outliers:", outliers)
```

# EDA Feature Scaling

Feature scaling is a preprocessing technique that transforms feature values to a similar scale, ensuring all features contribute equally to the model

<ul>
  <li><strong>Label Encoding for Ordinal Data</strong></li>
  <ul>
    <li><strong>Definition:</strong> Label Encoding is a technique used to convert categorical variables into numerical values. It assigns a unique integer to each category in an ordinal feature, where the order or ranking of categories matters.</li>
    <li><strong>Use Case:</strong> Useful for ordinal data, where there is an inherent order among the categories (e.g., education levels like High School < Bachelor's < Master's).</li>
    <li><strong>Process:</strong>
      <ul>
        <li>Each unique category is assigned a numerical label based on its order.</li>
        <li>For example, for education levels:</li>
        <ul>
          <li>High School → 0</li>
          <li>Bachelor’s → 1</li>
          <li>Master’s → 2</li>
        </ul>
      </ul>
    </li>
  </ul>
</ul>


```python
# Implementation (using LabelEncoder from sklearn):

from sklearn.preprocessing import LabelEncoder

data = ['High School', 'Bachelor', 'Master']
encoder = LabelEncoder()
encoded_data = encoder.fit_transform(data)
print(encoded_data)  
```

Pros: Simple and efficient for ordinal data, preserves order.

Cons: Not suitable for non-ordinal data as it may misrepresent relationships.



<ul>
  <li><strong>One-Hot Encoding for Categorical Data</strong></li>
  <ul>
    <li><strong>Definition:</strong> One-Hot Encoding converts categorical variables into a binary matrix, where each unique category in a feature is represented as a separate binary column (0 or 1).</li>
    <li><strong>Use Case:</strong> Ideal for nominal (categorical) data, where there is no intrinsic order (e.g., color: Red, Green, Blue).</li>
    <li><strong>Process:</strong>
      <ul>
        <li>Each unique category is represented by a new binary column.</li>
        <li>For example, for colors:</li>
        <ul>
          <li>Red → [1, 0, 0]</li>
          <li>Green → [0, 1, 0]</li>
          <li>Blue → [0, 0, 1]</li>
        </ul>
      </ul>
    </li>
  </ul>
</ul>


```python
# Implementation (using OneHotEncoder from sklearn):

from sklearn.preprocessing import OneHotEncoder
import numpy as np

data = np.array(['Red', 'Green', 'Blue']).reshape(-1, 1)
encoder = OneHotEncoder(sparse=False)
encoded_data = encoder.fit_transform(data)
print(encoded_data)
```

Pros: Effective for non-ordinal categorical data, avoids implying order among categories.

Cons: Can lead to high-dimensional data if there are many categories, which can increase computation time.
